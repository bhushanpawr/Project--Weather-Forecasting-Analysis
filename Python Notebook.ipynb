```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('weather.csv', encoding='latin1')
data
```

```python
data.head()
```

# drop columns

```python
data=data.drop('Sr. No.', axis=1)
data=data.drop('Date2',axis=1)
data=data.drop('Date1',axis=1)
```

```python
data
```

# Duplicated rows

```python
# print the duplicated rows
data[data.duplicated()]
```

```python
# drop duplicates
data=data.drop_duplicates()
```

```python
data
```

# Renaming the columns

```python
data.info()
```

```python
data = data.rename(columns={' Average humidity (%': 'Average_humidity_%',
                            ' Average dewpoint (°F': 'Average_dewpoint_°F',
                            ' Average barometer (in': 'Average_barometer_in',
                            ' Average windspeed (mph': 'Average_windspeed_mph',
                            ' Average gustspeed (mph': 'Average_gustspeed_mph',
                            ' Average direction (°deg': 'Average_direction_°deg',
                            ' Rainfall for month (in': 'Rainfall_for_month_in',
                            'Rainfall for year (in': 'Rainfall_for_year_in',
                            ' Maximum rain per minute': 'Maximum_rain_per_minute',
                            ' Maximum temperature (°F': 'Maximum_temperature_°F',
                            ' Minimum temperature (°F': 'Minimum_temperature_°F',
                            ' Maximum humidity (%': 'Maximum_humidity_%',
                            ' Minimum humidity (%': 'Minimum_humidity_%',
                            ' Maximum pressure': 'Maximum_pressure',
                            ' Minimum pressure': 'Minimum_pressure',
                            ' Maximum windspeed (mph': 'Maximum_windspeed_mph',
                            ' Maximum gust speed (mph': 'Maximum_gust_speed_mph',
                            ' Maximum heat index (°F': 'Maximum_heat_index_°F',
                            ' Temperature': 'Temperature',
                            ' Month': 'Month',
                            ' diff_pressure': 'diff_pressure'
                           })
```

```python
data.columns
```

# Handling null values

```python
# Checking null values
data.isnull().sum()
```

# Replacing values

```python
# Replace
data=data.replace('0.11700000000000088;','0.11700000000000088')
data
```

```python
# Data types
data.info()
```

```python
# change data types
data['diff_pressure'] = data['diff_pressure'].astype('float64')
```
